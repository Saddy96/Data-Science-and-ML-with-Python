```python
#Removing the Car name column

df.drop('Car_Name', axis = 1, inplace= True)
```

```python
# To find the age of the car
df['current_year'] = 2021
df.head()
```

```python
df['no_year'] = df['current_year'] - df['Year']
```

```python
df.head()
```

```python
df.drop('current_year', axis = 1, inplace= True)
```

```python
#Droping Year column as we have calculated the age of the car/no_year
df.drop('Year', axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Dummy Encoding, One hot Encooding
final_df = pd.get_dummies(df, drop_first= True)
```

```python
final_df.head()
```

```python
final_df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = final_df.corr()

plt.figure(figsize = (10,12))
sns.heatmap(corr, annot=True, linecolor = 'White', linewidths= 0.8)
plt.show()
```

Feature and Target Variable

```python
X = final_df.iloc[:,1:]
y = final_df.iloc[:,0]
```

```python
X.shape
```

```python
y.shape
```

Data Split

```python
import sklearn

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

Save Data file

```python
final_df.to_csv('test.csv')
```

Feature importance

```python
# Feature Importance- Which feature is important

from sklearn.ensemble import ExtraTreesRegressor

fi_model = ExtraTreesRegressor()
fi_model.fit(X_train, y_train) 

print(fi_model.feature_importances_)
```

```python
feat_imp = pd.Series(fi_model.feature_importances_, index = X.columns)
```

```python
feat_imp.nlargest(5).plot(kind = 'barh')
plt.show()
```

Model Creation

```python
# Model 1 - Linear Regression
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predict on teh test data
y_pred = lr_model.predict(X_test)

#print errors
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test,y_pred))
print('MSE:', metrics.mean_squared_error(y_test,y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))

R2 = metrics.r2_score(y_test, y_pred)

print('R2: ', R2)
```

```python
# Model 2 - Decission Tree 
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)

# Predict on teh test data
y_pred1 = dt_model.predict(X_test)

#print errors
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test,y_pred1))
print('MSE:', metrics.mean_squared_error(y_test,y_pred1))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,y_pred1)))

R2 = metrics.r2_score(y_test, y_pred1)

print('R2: ', R2)
```

```python
# Model 3 - Random Forest
from sklearn.ensemble import RandomForestRegressor

regressor_model = RandomForestRegressor(n_estimators=100, random_state = 42)
regressor_model.fit(X_train, y_train)

# Predict on teh test data
y_pred2 = regressor_model.predict(X_test)

#print errors
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test,y_pred2))
print('MSE:', metrics.mean_squared_error(y_test,y_pred2))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,y_pred2)))

R2 = metrics.r2_score(y_test, y_pred2)

print('R2: ', R2)
```
