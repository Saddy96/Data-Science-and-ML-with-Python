# Matplotlib 3-d

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
iris = pd.read_csv('/content/Iris.csv')

iris
```

```python
for species,irissubset in iris.groupby('Species'):
  plt.scatter(irissubset['PetalLengthCm'], irissubset['PetalWidthCm'], label = species)

plt.legend()
plt.show()
```

```python

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

for species, irissubset in iris.groupby('Species'):
  ax.scatter(irissubset['PetalLengthCm'], irissubset['PetalWidthCm'],irissubset['SepalLengthCm'], label = species)
plt.legend()

ax.set_xlabel('PetalLength')
ax.set_ylabel('PetalWidth')
ax.set_zlabel('SepalLength')

plt.show()
```

# AREA Graph

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]

sleeping =[7,8,6,11,7]

eating = [2,3,4,3,2]

working =[7,8,7,2,2]

playing = [8,5,7,8,13]

plt.plot([],[],color='m', label='Sleeping', linewidth=5)

plt.plot([],[],color='c', label='Eating', linewidth=5)

#plt.plot([],[],color='r', label='Working', linewidth=5)

#plt.plot([],[],color='k', label='Playing', linewidth=5)

plt.stackplot(days, sleeping,eating,working,playing,colors=['m','c','r','k'])

plt.xlabel('x')

plt.ylabel('y')

plt.title('Stack Plot')

plt.legend()

plt.show()
```

# Subplots - Multiple Plots

```python
import numpy as np

import matplotlib.pyplot as plt

def f(t):
  return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)

t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(221)  #no of plots, row wise or column wise, 1 2 plot
#2 - Two plots - rows
#1 - column wise (use 2)
#The layout is organized in rows and columns, which are represented by the first and second argument.

#The third argument represents the index of the current plot

plt.plot(t1, f(t1), 'bo', t2, f(t2))

plt.subplot(222)

plt.plot(t2, np.cos(2*np.pi*t2))

plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

def f(t):
  return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)

t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(311)  #no of plots, row wise or column wise, 1 2 plot
#2 - Two plots
#1 - row wise or column wise (use 2)
plt.plot(t1, f(t1), 'bo', t2, f(t2))

plt.subplot(312)

plt.plot(t2, np.cos(2*np.pi*t2))

plt.subplot(313)
plt.plot(t1, f(t1), 'r*', t2, f(t2))


plt.show()
```
