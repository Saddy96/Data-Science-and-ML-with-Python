#Bi-variate Analysis

```python
# In above case we have done univariate analysis.

# When we r doing the analysis of 2 var together
```

```python
# 2D Scatter plot - realtionship between two variable
sns.set_style('whitegrid')
#Facetgrid - Plot between 2 var 
sns.FacetGrid(data =df, hue = 'status').map(plt.scatter, "age", "nodes").add_legend()

#Scatter plot can't be used if there is more than 2 features. canbe achived using pair plot
```

```python
# pair -plot is not that useful for the high dimensional data
sns.set_style('darkgrid')
sns.pairplot(df, hue = 'status', vars = ['age', 'year', 'nodes'], palette='gist_rainbow')
```

```python
sns.set_style('dark')
sns.pairplot(df, hue = 'status')
```

# Univariate

```python
#Histogram for numerical, how the data are distributed

# Distribution - how many value falls in a range

live = df.loc[df['status'] == 1]
dead = df.loc[df['status'] == 2]

plt.plot(live['age'], np.zeros_like(live['age']), 'o')
plt.plot(dead['age'], np.zeros_like(dead['age']), 'o')
```

```python
sns.FacetGrid(df, hue = 'status').map(sns.distplot, 'year'). add_legend()
```

```python
sns.histplot(df['age'])
```
