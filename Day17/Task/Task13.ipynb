## Exploratory Data Analysis (EDA) of Titanic Survival Problem

To do the same we will use the Pandas, Seaborn and Matplotlib library.

Dataset contains the details of the passengers who had boarded the ship.

Dataset: [Titanic Data set](https://www.kaggle.com/c/titanic/data?select=test.csv)

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

import warnings
```

```python
df = pd.read_csv('/content/train.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.columns
```

# Check Class Imbalance

```python
df['Survived'].value_counts()
```

# univariate Analysis

```python
sns.set_style('dark')

sns.countplot(x = df['Survived'])
```

# Null Check

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), yticklabels=False)
```

```python
sns.countplot(x = df['Sex'])
```

```python
sns.histplot(df['Age'])
```

```python
sns.distplot(df['Age'])
```

```python
sns.set_palette('Reds')
sns.set_style('ticks')
sns.distplot(df['Fare'])
```

```python
sns.countplot(df['Survived'], hue = df['Sex'], data = df)
```

```python
sns.countplot('Pclass', hue = 'Survived', data = df)
```

```python
sns.countplot('Parch', hue = 'Survived', data = df, color = 'green')
```

# Bivariate Analysis

```python
sns.set_style('whitegrid')
#Facetgrid - Plot between 2 var 
sns.FacetGrid(data =df, hue = 'Survived').map(plt.scatter, "Age", "Fare").add_legend()
```

```python
sns.boxplot(x = 'Survived', y = 'Age', data = df)
```

```python
sns.boxplot(x = 'Pclass', y = 'Age', data = df)
```

```python
df.head()
```

```python
sns.set_palette('magma')
sns.pairplot(df, hue = 'Survived')
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), annot =True, linewidth = 0.5)
```
